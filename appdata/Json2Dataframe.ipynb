```python
from pprint import pprint as pp

input_folder = "minecraft/"
file = "usercache.json"

with open(input_folder + file) as f:
    str_df = f.read()

pp(str_df)
```

```python
import pandas as pd
import ast

output_folder = "Saved CSV/"
df = ast.literal_eval(str_df)

df = pd.DataFrame.from_dict(df)
basename = file.split('.')[0]
df.to_csv(output_folder + basename + '.csv')
```

```python
def Save_CSV(input_folder = "minecraft/", file = "usercache.json"):
    import pandas as pd
    import ast

    with open(input_folder + file) as f:
        str_df = f.read()

    output_folder = "Saved CSV/"
    df = ast.literal_eval(str_df)

    df = pd.DataFrame.from_dict(df)
    basename = file.split('.')[0]
    
    df.to_csv(output_folder + basename + '.csv')
    return df
```

```python
Save_CSV()
```

```python
import os, json

json_files = [pos_json for pos_json in os.listdir(input_folder) if pos_json.endswith('.json')]
json_files
```

```python
for i in json_files:
    print('\n', f' ({i}) '.center(80, "-"), end='\n\n')
    try: pp(Save_CSV(input_folder = "minecraft/", file = i))
    except Exception as e: print(e)
```
