```python
from pprint import pprint as pp

input_folder = "minecraft/"
file = "usercache.json"

with open(input_folder + file) as f:
    str_df = f.read()

pp(str_df)
```

```python
import pandas as pd
import ast

output_folder = "Saved CSV/"
df = ast.literal_eval(str_df)

df = pd.DataFrame.from_dict(df)
basename = file.split('.')[0]
df.to_csv(output_folder + basename + '.csv')
```

```python
def Save_CSV(file = "minecraft/usercache.json"):
    import pandas as pd
    import ast, os

    try:
        with open(file, encoding= 'unicode_escape') as f:
            str_df = f.read()
    except:
        with open(file, encoding= 'utf-8') as f:
            str_df = f.read()
        
    output_folder = "Saved CSV/"
    df = ast.literal_eval(str_df)

    df = pd.DataFrame.from_dict(df)
    file = os.path.basename(file)
    file = file.split('.')[0]
    
    df.to_csv(output_folder + file + '.csv')
    return df
```

```python
Save_CSV()
```

```python
import os

json_files = [os.path.join(root, name)
    for root, dirs, files in os.walk(input_folder)
    for name in files
    if name.endswith((".json", ".xml"))
]

json_files
```

```python
for i in json_files:
    print('\n', f' ({i}) '.center(100, "-"), end='\n\n')
    try: pp(Save_CSV(file = i))
    except Exception as e: print(e)
```
